```python
plt.figure(figsize=(15, 5))
x=cat_counts.index
y=cat_counts
plt.bar(x,y)
plt.xticks(rotation = 90, fontsize = 12)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
x=cat_counts.index
y=cat_counts
plt.barh(x,y)
plt.xticks(rotation = 90, fontsize = 12)
plt.show()
```
