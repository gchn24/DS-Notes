````python
df=pd.read_csv("final_vg.csv")
df.head()
````

````python
ih=df.loc[df["Name"]=="Ice Hockey"]
ih.head()
````

````python
ba=df.loc[df["Name"]=="Baseball"]
ba.head()
````

````python
top_plat=df["Platform"].value_counts().index[:3]
top_plat
````

````python
top_genre=df["Genre"].value_counts().index[:3]
top_genre
````

````python
top_pub=df["Publisher"].value_counts().index[:3]
top_pub
````

````python
df_final=df.loc[(df["Platform"].isin(top_plat)) & (df["Genre"].isin(top_genre)) & (df["Publisher"].isin(top_pub))]
df_final.head()
````

````python
fig=plt.figure(figsize=(12,10))

plt.subplot(2,3,1)

sns.scatterplot(data=df_final,
               x="NA_Sales",
               y="Global_Sales",)
plt.title("NA_Sales v/s Global_Sales")
 
plt.subplot(2,3,2)
sns.lineplot(data=df_final,
               x="Year",
               y="NA_Sales",)
plt.title("Year v/s NA_Sales")
    
plt.subplot(2,3,3)
sns.scatterplot(data=df_final,
               x="EU_Sales",
               y="Global_Sales",)
plt.title("EU_Sales v/s Global_Sales")

plt.subplot(2,3,4)
sns.lineplot(data=df_final,
               x="Year",
               y="EU_Sales",)
plt.title("EU_Sales v/s Year")

plt.subplot(2,3,5)
sns.scatterplot(data=df_final,
               x="Other_Sales",
               y="Global_Sales",)
plt.title("Other_Sales v/s Global_Sales")

plt.subplot(2,3,6)
sns.lineplot(data=df_final,
               x="Year",
               y="Other_Sales",)
plt.title("Other_Sales v/s Year")

fig.suptitle("Data about sales figures")

plt.show()
````

````python
fig=plt.figure(figsize=(15,10))

plt.subplot(2,3,1)
sns.scatterplot(data=df_final,
               x="NA_Sales",
               y="Global_Sales",
               hue="Genre")
plt.title("NA_Sales v/s Global_Sales")
 
plt.subplot(2,3,2)
sns.lineplot(data=df_final,
               x="Year",
               y="NA_Sales",
               hue="Genre")
plt.title("Year v/s NA_Sales")
    
plt.subplot(2,3,3)
sns.scatterplot(data=df_final,
               x="EU_Sales",
               y="Global_Sales",
               hue="Genre")
plt.title("EU_Sales v/s Global_Sales")

plt.subplot(2,3,4)
sns.lineplot(data=df_final,
               x="Year",
               y="EU_Sales",
               hue="Genre")
plt.title("EU_Sales v/s Year")

plt.subplot(2,3,5)
sns.scatterplot(data=df_final,
               x="Other_Sales",
               y="Global_Sales",
               hue="Genre")
plt.title("Other_Sales v/s Global_Sales")

plt.subplot(2,3,6)
sns.lineplot(data=df_final,
               x="Year",
               y="Other_Sales",
               hue="Genre")
plt.title("Other_Sales v/s Year")

fig.suptitle("Data about sales figures")

plt.show()
````
